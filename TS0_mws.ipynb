La consigna era simple, desarrollar un codigo que permita crear un generador de funciones sinusoidales teniendo en cuenta algunos parametros de entrada que permiten modificar ciertos aspectos de la funcion generada. Usando la libreria numpy de Python, en especifico el comando np.sin, que genera la señal sinusoidal. Teniendo eso en cuenta se creo una funcion cuyos parametros de entrada fueron denominados: A0, offset, fx, phase, nn, fs. Siendo A0 la amplitud maxima de la senoidal medida en volts, offset mueve la senoidal hacia arriba o hacia abajo en el eje y, fx determina la frecuencia de la señal, phase es la fase de la misma, nn es la cantidad de muestras que se van a tomar para generar la señal y finalmente fs es la frecuencia de muestreo medida en Hz, es decir me indica cuantas muestras por segundo se van a tomar. 

# Codigo

Teniendo en cuenta lo mencionado en la introduccion escribi el siguiente codigo.

Aclaracion: los valores asignados a las variables son los que discutimos en clase y fijamos como estandar.

```python
import matplotlib.pyplot as plt
import numpy as np

def mi_funcion_sen (A0, offset, fx, phase, nn, fs):
    tiempo = np.arange(0, N * Ts, Ts)
    x = A0 * np.sin(2 * np.pi * fx * tiempo)
    return tiempo, x


tiempo, x = mi_funcion_sen(A0 = 1, offset = 0, fx = 1, phase = 0, nn = N, fs = fs)
plt.figure(figsize=(8,4))
plt.plot(tiempo, x, label='Señal senoidal')
plt.title('Señal Senoidal')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [v]')
plt.grid(True)
plt.legend()
plt.show()
```

Como se puede ver en el grafico de arriba, el generador de senoidales funciona correctamente y crea la señal deseada a partir de los parametros propuestos. Un detalle que vale la pena resaltar es el Teorema de Nyquist-Shannon encargado de definir la frecuencia de visualizacion ideal de la señal, fuera de estas frecuencias empiezan a pasar cosas raras como las que veremos a continuacion. La frecuencia de Nyquist es:
$$f_{Nyquist}=\frac{f_s}{2}$$ 
En este caso como la frecuencia de muestreo es 1000, la maxima frecuencia que puedo usar sin ver errores (alising) es de 500 Hz.

# Analisis de casos particulares

A continuacion analizaremos que ocurre al varias la fx, es decir la frecuencia de la señal.
